## I. The case of a quadratic functional
<a id='I'></a> 
We first apply the algorithm to a quadratic functional:
$$ (1)\qquad f(x)=\dfrac12 x^T A x + b^T x +c,\qquad \text{for } x\in \mathbb{R}^N,$$
with
$A$ real, positive definite symmetric $N\times N$ matrix, $b\in\mathbb{R}^N$ and $c\in \mathbb{R}$. 
Hence, $f$ is strongly $\gamma$-convex for some $\gamma>0$, and has a unique minimizer $x^*$ characterized by the condition $\nabla f(x^*)=0$.

### The optimal step gradient descent algorithm 

Let $f$ be a convex and coercive function of class $C^1$ on $\mathbb{R}^N$. The optimal gradient descent algorithm is defined as follows. Let $x^0\in \mathbb{R}^N$ (we try to choose $x^0$ close to $x^*$, if no hint, then we take $x^0=0$). Then, for $k=0,1,2,\ldots\ $ up to convergence, repeat:

$$
\left|
\begin{array}{lcl}
v^k& \longleftarrow & -\nabla f(x^k),\\
\alpha^k &\longleftarrow &\mathop{argmin}_{t>0} f(x^k + tv^k),\\ 
x^{k+1}&\longleftarrow &x^k+\alpha^k v^k
\end{array}
\right.
$$

 <a id='q4'></a>
**Stopping criterion.** We will use the stopping criterion $\Vert \nabla f(x^k)\Vert\leq \textit{tol}\,\Vert\nabla f(x^0)\Vert$.

**Remark:** In general, we do not know how to compute $\alpha^k$ and in practice, the second step is replaced by an approximate line search. However, when $f$ is quadratic, the computation of $\alpha^k$ is easy.

$$
f(x) = \dfrac12 x^T A x + b^T x + c
$$

$$
\nabla f(x^k) = A x^k + b
$$

$$
v^k = - \nabla f(x^k) = - (A x^k + b)
$$





$$
f(x^k + t v^k)
= \dfrac12 (x^k + t v^k)^T A (x^k + t v^k)
+ b^T (x^k + t v^k) + c
$$

$$
f(x^k + t v^k)
= \dfrac12 \Big[
(x^k)^T A x^k
+ t (x^k)^T A v^k
+ t (v^k)^T A x^k
+ t^2 (v^k)^T A v^k
\Big]
+ b^T x^k
+ t b^T v^k
+ c
$$

$$
f(x^k + t v^k)
= \dfrac12 (v^k)^T A v^k \, t^2
+ \left(
\dfrac12 (x^k)^T A v^k
+ \dfrac12 (v^k)^T A x^k
+ b^T v^k
\right) t
+ \dfrac12 (x^k)^T A x^k
+ b^T x^k
+ c
$$

$$
\alpha^k = \arg\min_{t>0} f(x^k + t v^k)
$$

$$
\iff \dfrac{d}{dt} f(x^k + t v^k) = 0
$$

$$
\dfrac{d}{dt} f(x^k + t v^k)
= (v^k)^T A v^k \, t
+ \dfrac12 (x^k)^T A v^k
+ \dfrac12 (v^k)^T A x^k
+ b^T v^k
$$

$$
(v^k)^T A v^k \, t
+ \dfrac12 (x^k)^T A v^k
+ \dfrac12 (v^k)^T A x^k
+ b^T v^k
= 0
$$

$$
(v^k)^T A v^k \, t
+ (v^k)^T A x^k
+ b^T v^k
= 0
$$

$$
(v^k)^T A v^k \, t
+ (v^k)^T (A x^k + b)
= 0
$$

$$
t
= - \dfrac{(v^k)^T (A x^k + b)}{(v^k)^T A v^k}
$$

$$
(v^k)^T (A x^k + b)
= - \|A x^k + b\|^2
$$

$$
(v^k)^T A v^k
= (A x^k + b)^T A (A x^k + b)
$$

$$
\boxed{
\alpha^k
= \dfrac{\|A x^k + b\|^2}
{(A x^k + b)^T A (A x^k + b)}
}
$$

Two functions are given below: 
- a function that draws a vector field given by a mapping $F$. As an example it is applied to the mapping $G(x,y)=(x, 9y)$.
- a functions that draws some level lines of a function $f$. It is applied to 
$g(x,y)=\dfrac{x^2+9y^2}2$. Notice that $G=\nabla g$. What do you observe?

```python
import numpy as np
import matplotlib.pyplot as plt 

def draw_vector_field(F, xmin, xmax, ymin, ymax, N=15):
    X = np.linspace(xmin, xmax, N)  # x coordinates of the grid points
    Y = np.linspace(ymin, ymax, N)  # y coordinates of the grid points
    U, V = F(*np.meshgrid(X, Y))  # vector field
    M = np.hypot(U, V)  # compute the norm of (U,V)
    M[M == 0] = 1  # avoid division by 0
    U /= M  # normalize the u componant
    V /= M  # normalize the v componant
    return plt.quiver(X, Y, U, V, angles='xy')

def level_lines(f, xmin, xmax, ymin, ymax, levels, N=500):
    x = np.linspace(xmin, xmax, N)
    y = np.linspace(ymin, ymax, N)
    z = f(*np.meshgrid(x, y))
    level_l = plt.contour(x, y, z, levels=levels)
    #plt.clabel(level_l, levels, fmt='%.1f') 


g = lambda x, y: .5*(x**2 + 9*y**2)
G = lambda x, y: np.array([x, 9*y])

level_lines(g, -5, 5, -2, 2, np.linspace(0, 12, 4))
draw_vector_field(G,  -5, 5, -2, 2, 10)
plt.axis('equal')
plt.show()
```

**Question 3.**  Implement the optimal gradient descent algorithm below, along with a graph representation of the iterations. The starting point should be $x^0=\binom{1}{C}$. On the same graphics,  using the functions above, represent the iterations, some level sets of $f$ and the normalized vector field $\dfrac {1}{|\nabla f|}\nabla f$.
  


```python
## solution 3
## initialization: make_f(c) returns the function with C=c,  make_df(c) returns its gradient function
def make_f(c):
    return lambda x, y=1: .5*(c*x**2 + y**2)
##


def make_df(c): 
    return lambda x, y=1: np.array([c*x, y])

## Setting tolerance, max number of iterations, and defining the function and its gradient
tol = 1e-8
max_niter = 1000
C = 16.
f, df =make_f(C), make_df(C)


## Setting initial point x0 = (x,y), initial value f(x0)=fX, initial gradient value gX=df(x,y) and its norm n
x, y = 1., C
fX, gX =f(x,y), df(x,y)
n=np.linalg.norm(gX)
tau = n*tol

## Initialisation of the iterative process: k = current step, X = array for the values of $x^k$ so far,
##                                          F=array for the values of $f(x^k)$ so far, n = array of values on gradient norms so far
k=0
X = [np.array([x,y])] # array for the values of $x^k$
F = [fX] # array for the values of $f(x^k)$
nG = [n]
```

```python
## Optimization loop: Fill in the BLANKS
while nG[k] > tol * nG[0] :
    alpha = (gX @ gX) / (gX @ np.array([[C, 0.],[0., 1.]]) @ gX)
    x = x - alpha*gX[0]
    y = y - alpha*gX[1]
    fX = f(x,y)
    gX = df(x,y)
    n = np.linalg.norm(gX)
    
    k+= 1
    X.append(np.array([x,y]))
    F.append(fX)
    nG.append(n)
    
X = np.array(X)
F = np.array(F)
nG =np.array(nG)
print(f"niter ={k}")
```

```python
# plot the results 
plt.figure()
plt.plot(X[:,0],X[:,1],'.',linestyle='-')
level_lines(f, -2.1, 2.1, 0, C, np.linspace(0, .5*C**2, 10))
draw_vector_field(df,  -2.1, 2.1, 0, C, 20)
plt.axis('equal')

# plot of the values of f along the iterations.
plt.figure()
plt.semilogy(range(len(F)),F,'.',linestyle='dashed')

plt.show()
```

**Question 4.** Change the value of $C$ from $1, 2, 4, 8, 16, 32$. What do you observe?

```python
## solution 4: Fill in the BLANK for the optimization loop
def trace_trajectory(C=2):

    ##Initialization
    tol = 1e-8
    max_niter = 1000
    f, df =make_f(C), make_df(C)    
    x, y = 1., C
    fX, gX =f(x,y), df(x,y)
    n=np.linalg.norm(gX)
    tau = n*tol
    k=0
    X = [np.array([x,y])] # array for the values of $x^k$
    F = [fX] # array for the values of $f(x^k)$
    nG = [n]

    ## Optimization loop: Fill in the BLANKS
    while nG[k] > tol * nG[0] :
        alpha = (gX @ gX) / (gX @ np.array([[C, 0.],[0., 1.]]) @ gX)
        x = x - alpha*gX[0]
        y = y - alpha*gX[1]
        fX = f(x,y)
        gX = df(x,y)
        n = np.linalg.norm(gX)
        
        k+= 1
        X.append(np.array([x,y]))
        F.append(fX)
        nG.append(n)
        
    X = np.array(X)
    F = np.array(F)
    nG =np.array(nG)
    print(f"niter ={k}")

    ## plot the results 
    plt.figure()
    plt.plot(X[:,0],X[:,1],'.',linestyle='-')
    level_lines(f, -2.1, 2.1, 0, C, np.linspace(0, .5*C**2, 10))
    draw_vector_field(df,-2.1, 2.1, 0, C, 15)
    plt.axis('equal')

    ## plot of the values of f along the iterations in a log-scale.
    plt.figure()
    plt.semilogy(range(len(F)),F,'.',linestyle='dashed')
    plt.show()
    
for C in (1,2,4,8,16,32):
    trace_trajectory(C)
    
```

**Question 7.** Make an hypothesis explaining these results. 

___Solution 7___

For the quadratic function
$
f(x)=\frac12 x^T A x,
\qquad 
A=\begin{pmatrix}1&0\\0&C\end{pmatrix}, \quad C\ge 1,
$
the eigenvalues of \(A\) are
$
\lambda_{\min}=1, \qquad \lambda_{\max}=C.
$
The condition number of \(A\) is therefore
$
\kappa(A)=\frac{\lambda_{\max}}{\lambda_{\min}}=C.
$

As \(C\) increases, the problem becomes increasingly ill-conditioned.  
The level sets of \(f\) are more and more elongated ellipses, which causes the gradient descent method to exhibit a zig-zag behavior.

For quadratic functions with exact line search, the convergence factor of the gradient descent method is
$
\rho = \frac{\kappa(A)-1}{\kappa(A)+1}.
$
Thus, when \(C\) increases, \(\rho\) approaches \(1\), and the convergence becomes slower.  
Consequently, the number of iterations required to reach a given tolerance increases with \(C\).


## II. The case of a smooth functional, line search  <a id='II'></a>

We consider the function defined as
$$
f(x,y):= \cosh(x) + \sin^2(x+y),\qquad \text{for }z=(x,y)\in \mathbb{R}^2.
$$

 <a id='q11'></a>
**Question 8.** Show that $\ \inf_{\mathbb{R}^2} f\ = 1$ and that the minimizers of $f$ in $\mathbb{R}^2$ are the point $z^n_*:=(0,n\pi)$, $n\in\mathbb{Z}$.

___Solution 8___

### $$\ \inf_{\mathbb{R}^2} f\ = 1 $$

we have that $ cosh(x) = (\exp(x) + \exp(-x))/2 >= 1 $
and $ \sin^2(x+y) >= 0 $
thus,  $$ f(x,y) >= 1  =>\ \inf_{\mathbb{R}^2} f\ = 1 $$

### the minimizers of $f$ in $\mathbb{R}^2$ are the point $z^n_*:=(0,n\pi)$
$$ f(x,y):= \cosh(x) + \sin^2(x+y)  = (\exp(x) + \exp(-x))/2 + \sin^2(x+y) $$
$$ df(x,y)/dx = (\exp(x) - \exp(-x))/2 + 2*sin(x+y)*cos(x+y) = (\exp(x) - \exp(-x))/2 + sin(2*(x+y))$$
$$ df(x,y)/dy =  2*sin(x+y)*cos(x+y)  = sin(2*(x+y)))$$
$$ \nabla f(x,y) = ((\exp(x) - \exp(-x))/2 + 2*sin(x+y)*cos(x+y) , 2*sin(x+y)*cos(x+y) ) $$
The mimimize of f(x,y) is given by $$ \nabla f(x,y) = 0  $$ 
since $$ \nabla f(z)  = \nabla f(0,n*\pi)  = (0 + 2*sin(n*\pi)*cos(n*\pi) , 2*sin(n*\pi)*cos(n*\pi) ) = (0,0)$$
then z are the minimizers of f

**Question 9.** Show that $f$ is convex in the neighborhood of $z^0_*$.

___Solution 9___

$$ Hf(x,y) = \begin{pmatrix}
 (\exp(x) + \exp(-x))/2 + 2*cos(2*(x+y)) &  2*cos(2*(x+y)) \\
 2*cos(2*(x+y)) & 2*cos(2*(x+y))
\end{pmatrix} $$

$$ Hf(z^0_*)  = Hf(0,0) = \begin{pmatrix}
 3 &  2 \\
 2 & 2
\end{pmatrix} $$

Let $ v \in \mathbb{R}^2  ,  v = (a,b)$

$$ v.T * Hf(0,0) * v = 3*a + 4 a*b + 2*b = a + 2(a + 2a*b + b) = a + 2*(a+b)  >= 0$$ 

# So f is convex in the neighborhood of $z^0_*$.

We are going to apply a gradient descent algorithm with line search to the function $f$. Namely: given $z^0=(x^0,y^0)\in\mathbb{R}^2$, compute recursively, up to convergence: 
$$
\left|
\begin{array}{lcl}
v^k& \longleftarrow & -\nabla f(z^k),\\
\alpha^k &\longleftarrow & \text{Line-search}\ \left(\ t\mapsto f(z^k + tv^k)\ \right),\\ 
z^{k+1}&\longleftarrow &z^k+\alpha^k v^k
\end{array}
\right.
$$

Let us precise the second step. We denote 
$$
h_k(t):= f(z^k+tv^k).
$$
First we notice that 
$$
m_k:=h_k'(0) = \nabla(f(z^k))\cdot d_k=-\|\nabla f(z^k)\|^2<0
$$
and for $t>0$,
$$
(TE)\quad f(z^k+t v^k) \,=\, f(z^k) + m_k t +o(t),
$$
with 
$$
m_k<0.
$$
In fact, if $f$ is convex in the neighborhood of $z^k$, we also have for $t>0$ small enough,
$$
f(z^k+t v^k)\, \ge\, f(z^k) +m_kt ,
$$
so we can not ask for $f(z^k+t v^k) \,\le\, f(z^k) + m_k t $. The idea of the *Armijo condition* is to ask for a little less. We fix some $\delta\in (0,1)$: the Armijo condition reads:
$$
(AC)\quad f(z^k+t v^k)\, \le\, f(z^k) +\delta\,m_k\, t.
$$
Thanks to \eqref{TE} We know that \eqref{AC} holds true for $t>0$ small enough, but we do not want to choose too small an $\alpha^k$ (the algorithm would stall). So, we fix $\beta\in(0,1)$ and we start with some "large" $t=\alpha_0$. If \eqref{AC} holds true for this $t$, we pick $\alpha^k=t$. If this is wrong, we set $t\leftarrow \beta t$ and repeat the process until \eqref{AC} holds true (this will happen eventually thanks to the Taylor expansion \eqref{TE}). In the end, we set $\alpha^k=t$.

**Question 10.** Implement the above method, with $\delta=0.5$, $\beta=0.75$ and $\alpha_0=1$. Start with $z^0=(1,0.5)$. Plot the results.

First to help you, the next cell shows some level sets of $f$ and the normalized vector field $\dfrac {1}{|\nabla f|}\nabla f$ in the neighborhood of $z^*$.

```python
f = lambda x, y : np.cosh(x)+ np.sin(x + y)**2
df = lambda x, y : np.array(
    [np.sinh(x) + 2*np.cos(x + y)*np.sin(x + y),
     2*np.cos(x + y)*np.sin(x + y)])

level_lines(f, -1.1, 1.1, -1.1, 1.1, 
            np.linspace(1, 3, 10))
draw_vector_field(df, -1.1, 1.1, -1.1, 1.1, 10)
plt.axis('equal')
plt.show()
```

```python
## Solution 10: 

## Parameters
tol = 1e-6
delta, beta = 0.5, 0.75
alpha0 = 1

## Fixing max iteration number for both descent and linesearch
niter_max = 200
niter_ls_max = 20

## initialization 
k=0
z = np.array([1., .5]) 
fz = f(z[0],z[1])
gz = df(z[0], z[1])
ngz = np.hypot(gz[0], gz[1])
tau = tol*ngz
Z, F, NG =[z], [fz], [ngz]
flag = 'OK'
```

```python
## Optmization loop
while ngz > tol and k < niter_max:
    v = -gz                       
    mk = -ngz**2                  
    t = alpha0
    for _ in range(niter_ls_max):
        z_trial = z + t * v
        if f(z_trial[0], z_trial[1]) <= fz + delta * mk * t:
            break
        t *= beta
    else:
        flag = "Line search failed"
        break
        
    z = z + t * v
    fz = f(z[0], z[1])
    gz = df(z[0], z[1])
    ngz = np.linalg.norm(gz)

    
    k+=1 
    Z.append(z)
    F.append(fz)
    NG.append(ngz)
print(f"flag = {flag}, n_iter ={k}")   
```

```python
## Solution 10 - b

Z = np.array(Z)
F = np.array(F)
NG = np.array(NG)

plt.figure()
plt.plot(Z[:,0],Z[:,1],'.',linestyle='-')
level_lines(f, -1.1, 1.1, -1.1, 1.1, 
            np.linspace(1, 3, 10))
draw_vector_field(df, -1.1, 1.1, -1.1, 1.1, 10)
plt.axis('equal')

# plot of the values of f along the iterations.
plt.figure()
plt.semilogy(range(len(F)),F,'.',linestyle='dashed')

plt.show()
```

We consider now the fonction defined on $\mathbb{R}^3$ by 
$$
f(x,y,z):= \cosh(x) + \sin^2(x+y) + (y-z)^2,\qquad \text{for }w=(x,y,z)\in \mathbb{R}^2.
$$

 <a id='q14'></a>
**Question 11.** Apply the above optimization method to this function, starting from $w^0=(1,0.5,1)$.  

```python
f = lambda w : np.cosh(w[0])+ np.sin(
    w[0] + w[1])**2 + (w[1] - w[2])**2
df = lambda w : np.array(
     [np.sinh(w[0])+ 2*np.cos(w[0] + w[1])
      *np.sin(w[0] + w[1]), 
      2*np.cos(w[0] + w[1])*np.sin(w[0] + w[1])
      + 2*(w[1] - w[2]),  2*(w[2] - w[1])])
```

```python
## Solution 11

# Parameters
tol = 1e-6
delta, beta = 0.5, 0.75
alpha0 = 1.0

niter_max = 200
niter_ls_max = 20

# Initialization
k = 0
z = np.array([1.0, 0.5, 1.0])
fz = f(z)
gz = df(z)
ngz = np.linalg.norm(gz)

Z = [z.copy()]
F = [fz]
NG = [ngz]

flag = "OK"

# Optimization loop
while ngz > tol and k < niter_max:
    v = -gz
    mk = -ngz**2
    t = alpha0
    for _ in range(niter_ls_max):
        z_trial = z + t * v
        if f(z_trial) <= fz + delta * mk * t:
            break
        t *= beta
    else:
        flag = "Line search failed"
        break
    z = z + t * v
    fz = f(z)
    gz = df(z)
    ngz = np.linalg.norm(gz)

    Z.append(z)
    F.append(fz)
    NG.append(ngz)

    k += 1

print(f"flag = {flag}, n_iter = {k}")
```

## 3. Newton's method
 <a id='III'></a> 

### Newton's method for continuous optimization

In the context of continous optimization, Newton's method consists in approximating the function $f$ by its second order Taylor expansion around the current point $x^k$.
$$
f(x)=f(x^k)+\nabla f(x^k)\cdot(x-x^k)+ \dfrac12 (x-x^k)^TH_k(x-x^k),
$$
where $H^k:=d\nabla f(x^k)$ is the Hessian matrix of $f$ at point $x^k$.
$$
H^k_{ij}=\dfrac{\partial^2}{\partial x_i\partial x_j}f(x^k).
$$
We recall that $H^k$ is symmetric. Assuming furthermore that $f$ is $\gamma$-convex ($\gamma>0$), we have $H^k\ge\delta\, \mathrm{Id}$, that is
$$
v^TH^kv\ge \delta \Vert v\Vert^2\quad\forall\,y\in\mathbb{R}^N.
$$
In particular, $H^k$ is invertible. 

In the sequel, we assume that $f$ is $C^2$ and $\delta$-convex.

From class, we know the minimizer $v^k$ of 
$$
q_k(v):=f(x^k)+\nabla f(x^k)\cdot v+ \dfrac12 v^TH^kv
$$
is $v^k=-(H^k)^{-1}\nabla f(x^k)$. 

Then $v^k$ is a descent direction, that is 
$$\nabla f(x^k)\cdot d^k<0,$$
whenever $\nabla f(x^k)\ne0$.

A first algorithm, would consist in : for $k=0$ up to convergence, 
$$
\left|
\begin{array}{lcl}
v^k& \longleftarrow & -(H^k)^{-1}\nabla f(x^k),\\
x^{k+1}&\longleftarrow &x^k+ v^k.
\end{array}
\right.
$$

To be efficient the pure Newton's method requires the first iterate $x^0$ to be close to a solution $x^*$ (here the minimizer). For this reason, it is safer to use $d^k$ as a descent direction and apply a Line-search. This leads to 
$$
\left|
\begin{array}{lcl}
v^k& \longleftarrow & \text{computed  by your formula},\\
\alpha^k &\longleftarrow & \text{Line-search}\ \left(\ t\mapsto f(x^k + tv^k)\ \right),\\ 
x^{k+1}&\longleftarrow &x^k+\alpha^k v^k
\end{array}
\right.
$$
__!!! It is very important here to start the line search by first testing $t=1$ ($\alpha_0=1$) so that $\alpha^k=1$ when $x^k$ is close enough to $x^*$ and you recover the quadratic convergence of Newton's method. !!!__

Last tip: Take a small $\delta$, for instance $\delta=1/10$ so that the Armijo condition is not very demanding.

**Question 12.**  <a id='q17'></a>
Implement this method for the optimization of 
$$
f(x,y):= \cosh(x) + \sin^2(x+y).
$$
Observe numerically the rate of convergence.

```python
f = lambda x, y : np.cosh(x)+ np.sin(x + y)**2
df = lambda x, y : np.array([np.sinh(x) 
             + 2*np.cos(x + y)*np.sin(x + y), 
             2*np.cos(x + y)*np.sin(x + y)])
def ddf_inverse(x,y):
    a,ah=1/(2*np.cos(2*(x+y))), 1/np.cosh(x)
    return np.array([[ah,-ah],[-ah,a+ah]])
```

```python
## solution 12
tol = 1e-6
delta, beta = 0.1, 0.75
alpha0 = 1.0

niter_max = 200
niter_ls_max = 20

# Initialization
k=0
z = np.array([1., .5]) 
fz = f(z[0],z[1])
gz = df(z[0], z[1])
ngz = np.hypot(gz[0], gz[1])
tau = tol*ngz
Z, F, NG =[z], [fz], [ngz]
flag = 'OK'

# Optimization loop
while ngz > tol and k < niter_max:
    x,y = z
    v = - ddf_inverse(x,y) @ gz
    mk = -ngz**2
    t = alpha0
    for _ in range(niter_ls_max):
        z_trial = z + t * v
        if f(z_trial[0],z_trial[1]) <= fz + delta * mk * t:
            break
        t *= beta
    else:
        flag = "Line search failed"
        break
    z = z + t * v
    fz = f(z[0] , z[1])
    gz = df(z[0],z[1])
    ngz = np.linalg.norm(gz)

    Z.append(z)
    F.append(fz)
    NG.append(ngz)

    k += 1

print(f"flag = {flag}, n_iter = {k}")
```
